### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
pd.set_option('display.max_rows', None)
df.isnull().sum()
```

```python
pd.reset_option('display.max_rows')
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(10,8))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# To check number of missing rows of a particular column
print("Number of missing rows in Employment: ",df['Employment'].isnull().sum())


pd.set_option('display.max_rows', None)
print("Number of mmissing rows in each column: \n")
missing_rows = df.isnull().sum()
missing_rows
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
freq_remote = df['RemoteWork'].value_counts().idxmax()

freq_edLevel = df["EdLevel"].value_counts().idxmax()

print(f"Most frequent value in RemoteWork column: {freq_remote}")
print(f"Most frequent value in EdLevel column: {freq_edLevel}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Imputing most freq values
df["RemoteWork"] = df["RemoteWork"].fillna(freq_remote)
df["EdLevel"] = df["EdLevel"].fillna(freq_edLevel)

print(f"After Imputing, Number of empty rows in RemoteWork and EdLevel columns respectively {df["RemoteWork"].isnull().sum()} and {df["EdLevel"].isnull().sum()}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
plt.figure(figsize=(10,6))
df["RemoteWork"].value_counts().plot(kind='bar',color='skyblue', edgecolor='black')
plt.xlabel('Remote Work')
plt.xticks(rotation=0)
plt.ylabel('Count')
plt.title("Distribution of values in RemoteWork")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df["EdLevel"], color='green')
plt.xticks(rotation=45, ha='right')
plt.title("EdLevel distribution after imputation")
plt.show()
```

```python
top_15 = df["Employment"].value_counts().head(20)
top_15

plt.figure(figsize=(10,6))
sns.barplot(x=top_15.index, y=top_15.values, color='orange')
plt.title('Data Distribution in Employment')
plt.xticks(rotation=90, ha='right')
plt.ylabel("Count")
plt.show()
```
