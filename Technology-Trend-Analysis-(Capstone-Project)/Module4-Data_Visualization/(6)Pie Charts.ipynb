**Download and Load the Data**


To start, download and load the dataset into a `pandas` DataFrame.



```python
# Step 1: Download the dataset
#!wget -O survey-data.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

# Step 2: Import necessary libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Display the first few rows to understand the structure of the data
pd.set_option('display.max_columns', None)  # Show all columns in the DataFrame
df.head()
```

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

# Step 1: Drop NaN values
df_db = df['DatabaseWantToWorkWith'].dropna()

# Step 2: Split semicolon-separated values and flatten the list
all_databases = df_db.str.split(';').explode().str.strip()

# Step 3: Count frequencies
db_counts = all_databases.value_counts()

# Step 4: Select top 5
top_5_dbs = db_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_dbs,
    labels=top_5_dbs.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here

# Step 1: Drop missing values
df_dev = df['DevType'].dropna()

# Step 2: Split multiple roles and flatten
all_dev_roles = df_dev.str.split(';').explode().str.strip()

# Step 3: Count frequency of each developer role
dev_role_counts = all_dev_roles.value_counts()

# Step 4: Get top 5 roles
top_5_roles = dev_role_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_roles,
    labels=top_5_roles.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel1.colors
)
plt.title('Top 5 Developer Roles Among Respondents')
plt.axis('equal')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here

# Step 1: Drop missing values
df_os = df['OpSysProfessional use'].dropna()

# Step 2: Split multi-OS entries and normalize
all_os = df_os.str.split(';').explode().str.strip()

# Step 3: Count OS usage
os_counts = all_os.value_counts()

# Step 4: Select top 5 most used OS
top_5_os = os_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_os,
    labels=top_5_os.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel2.colors
)
plt.title('Top 5 Operating Systems Used Professionally')
plt.axis('equal')
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here

# Step 1: Drop missing values
df_lang = df['LanguageHaveWorkedWith'].dropna()

# Step 2: Split multiple languages per row and normalize
all_langs = df_lang.str.split(';').explode().str.strip()

# Step 3: Count language frequency
lang_counts = all_langs.value_counts()

# Step 4: Select top 5 languages
top_5_langs = lang_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_langs,
    labels=top_5_langs.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Programming Languages Respondents Have Worked With')
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here

# Step 1: Drop missing values
df_tools = df['NEWCollabToolsHaveWorkedWith'].dropna()

# Step 2: Split each entry and normalize
all_tools = df_tools.str.split(';').explode().str.strip()

# Step 3: Count individual tool frequency
tool_counts = all_tools.value_counts()

# Step 4: Select top 5 tools
top_5_tools = tool_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_tools,
    labels=top_5_tools.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel2.colors
)
plt.title('Top 5 Collaboration Tools Used in Professional Work')
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here

# Step 1: Drop missing values
df_lang = df['LanguageAdmired'].dropna()

# Step 2: Split each entry by ';', then explode into rows
all_languages = df_lang.str.split(';').explode().str.strip()

# Step 3: Count individual language frequency
lang_counts = all_languages.value_counts()

# Step 4: Get top 5 languages
top_5_langs = lang_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_langs,
    labels=top_5_langs.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Admired Programming Languages')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.


```python
##Write your code here
# Step 1: Drop missing values
df_ai = df['AIToolCurrently Using'].dropna()

# Step 2: Split by ';' and explode into separate rows
ai_tools = df_ai.str.split(';').explode().str.strip()

# Step 3: Count frequency of each unique AI tool/task
ai_counts = ai_tools.value_counts()

# Step 4: Get top 5 AI tasks/tools
top_5_ai = ai_counts.head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_ai,
    labels=top_5_ai.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Top 5 AI Tools/Tasks Developers Are Currently Using')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here

# Step 1: Drop missing values from the column
df_web = df['WebframeWantToWorkWith'].dropna()

# Step 2: Split semicolon-separated frameworks and flatten
frameworks = df_web.str.split(';').explode().str.strip()

# Step 3: Count frequency of each individual framework
framework_counts = frameworks.value_counts()

# Step 4: Select top 5 frameworks
top_5_frameworks = framework_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_frameworks,
    labels=top_5_frameworks.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel1.colors
)
plt.title('Top 5 Web Frameworks Developers Want to Work With')
plt.axis('equal')  # Ensures the pie is drawn as a circle
plt.show()
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
##Write your code here

# Step 1: Drop missing values
df_embedded = df['EmbeddedWantToWorkWith'].dropna()

# Step 2: Split semicolon-separated values and explode to individual entries
embedded_tech = df_embedded.str.split(';').explode().str.strip()

# Step 3: Count occurrences of each embedded technology
embedded_counts = embedded_tech.value_counts()

# Step 4: Get top 5 technologies
top_5_embedded = embedded_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_5_embedded,
    labels=top_5_embedded.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)
plt.title('Top 5 Embedded Technologies Developers Want to Work With')
plt.axis('equal')  # Keep the pie chart circular
plt.show()
```
