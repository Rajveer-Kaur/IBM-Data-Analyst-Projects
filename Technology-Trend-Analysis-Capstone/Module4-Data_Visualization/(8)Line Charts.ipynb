### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here

# Filter out rows with missing compensation or age
df_agecomp = df[['Age', 'ConvertedCompYearly']].dropna()

# Optional: clean age groups for ordering
age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older',
    'Prefer not to say'
]

# Calculate median compensation per age group
median_comp_by_age = (
    df_agecomp.groupby('Age')['ConvertedCompYearly']
    .median()
    .reindex(age_order)
)

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', color='blue', linestyle='-')
plt.title('Median Yearly Compensation Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here

# Filter relevant age groups
age_focus = ['25-34 years old', '35-44 years old']
df_focus = df[df['Age'].isin(age_focus)]

# Drop missing compensation values
df_focus = df_focus[['Age', 'ConvertedCompYearly']].dropna()

# Compute median compensation
median_comp_focus = (
    df_focus.groupby('Age')['ConvertedCompYearly']
    .median()
    .reindex(age_focus)
)

# Plot line chart
plt.figure(figsize=(8, 5))
plt.plot(median_comp_focus.index, median_comp_focus.values, marker='o', linestyle='-', color='green')
plt.title('Median Compensation for Ages 25–45')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here

# Convert 'YearsCodePro' to numeric
#df['YearsCodePro_clean'] = df['YearsCodePro'].replace('Less than 1 year', '0').replace('More than 50 years', '51').astype(float)

# Drop missing values
#df_exp = df[['YearsCodePro_clean', 'JobSatPoints_6']].dropna()
df_exp = df[['WorkExp', 'JobSatPoints_6']].dropna()

# Group by experience and calculate average satisfaction
satisfaction_by_exp = df_exp.groupby('WorkExp')['JobSatPoints_6'].mean()

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(satisfaction_by_exp.index, satisfaction_by_exp.values, marker='o', color='blue')
plt.title('Job Satisfaction (JobSatPoints_6) by Years of Professional Coding Experience')
plt.xlabel('Years of Professional Experience')
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

# Drop rows with missing values in the two columns
df_exp = df[['WorkExp', 'ConvertedCompYearly']].dropna()

# Group by years of work experience and compute the median compensation
median_comp_by_exp = df_exp.groupby('WorkExp')['ConvertedCompYearly'].median()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(median_comp_by_exp.index, median_comp_by_exp.values, marker='o', linestyle='-', color='darkorange')
plt.title('Median Yearly Compensation vs. Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Median Yearly Compensation (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here

# Replace 'Less than 1 year' and other strings in YearsCodePro
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': '0',
    'More than 50 years': '51'  # if present
})

# Convert to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop missing values
df_exp = df[['YearsCodePro', 'JobSatPoints_6']].dropna()

# Group by experience and calculate median satisfaction
median_satisfaction = df_exp.groupby('YearsCodePro')['JobSatPoints_6'].median()

# Plot the line chart
plt.figure(figsize=(12, 6))
plt.plot(median_satisfaction.index, median_satisfaction.values, marker='o', color='royalblue')
plt.title('Median Job Satisfaction by Professional Coding Experience (YearsCodePro)')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
